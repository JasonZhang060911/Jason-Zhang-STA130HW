````python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import numpy as np

# Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Function to create a histogram with lines and rectangles
def add_annotations(fig, species_data, species_name):
    mean = species_data['flipper_length_mm'].mean()
    median = species_data['flipper_length_mm'].median()
    std = species_data['flipper_length_mm'].std()
    q1 = species_data['flipper_length_mm'].quantile(0.25)
    q3 = species_data['flipper_length_mm'].quantile(0.75)
    min_val = species_data['flipper_length_mm'].min()
    max_val = species_data['flipper_length_mm'].max()
    
    # Add vertical lines for mean and median
    fig.add_vline(x=mean, line=dict(color='blue', dash='dash'), annotation_text="Mean", row=species_name, col=1)
    fig.add_vline(x=median, line=dict(color='green', dash='dash'), annotation_text="Median", row=species_name, col=1)
    
    # Add rectangles for ranges (min-max), IQR, and +/- 2 standard deviations
    fig.add_vrect(x0=min_val, x1=max_val, fillcolor="lightgray", opacity=0.3, line_width=0, row=species_name, col=1, annotation_text="Range")
    fig.add_vrect(x0=q1, x1=q3, fillcolor="yellow", opacity=0.3, line_width=0, row=species_name, col=1, annotation_text="IQR")
    fig.add_vrect(x0=mean-2*std, x1=mean+2*std, fillcolor="red", opacity=0.2, line_width=0, row=species_name, col=1, annotation_text="+/- 2 STD")
    
# Create a subplot figure to hold histograms for each species
fig = make_subplots(rows=3, cols=1, subplot_titles=("Adelie", "Chinstrap", "Gentoo"))

# Create histograms for each species and add them to the figure
species_list = penguins['species'].unique()

for i, species in enumerate(species_list, 1):
    species_data = penguins[penguins['species'] == species]
    
    # Create a histogram for the species
    fig.add_trace(go.Histogram(x=species_data['flipper_length_mm'], name=species), row=i, col=1)
    
    # Add annotations (lines and rectangles)
    add_annotations(fig, species_data, i)

# Update layout
fig.update_layout(height=900, width=700, title_text="Flipper Length Distribution by Species with Annotations")

# Show the figure
fig.show(renderer="png")
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Function to plot KDE with annotations
def plot_kde_with_annotations(ax, species_data, species_name):
    # Plot KDE
    sns.kdeplot(species_data['flipper_length_mm'], ax=ax, fill=True, color="skyblue", label=f'{species_name} KDE')
    
    # Calculate statistics
    mean = species_data['flipper_length_mm'].mean()
    median = species_data['flipper_length_mm'].median()
    std = species_data['flipper_length_mm'].std()
    q1 = species_data['flipper_length_mm'].quantile(0.25)
    q3 = species_data['flipper_length_mm'].quantile(0.75)
    min_val = species_data['flipper_length_mm'].min()
    max_val = species_data['flipper_length_mm'].max()

    # Add vertical lines for mean and median
    ax.axvline(mean, color='blue', linestyle='--', label='Mean')
    ax.axvline(median, color='green', linestyle='--', label='Median')
    
    # Add shaded rectangles for ranges
    ax.axvspan(min_val, max_val, color='lightgray', alpha=0.3, label='Range (Min-Max)')
    ax.axvspan(q1, q3, color='yellow', alpha=0.3, label='IQR (Q1-Q3)')
    ax.axvspan(mean - 2*std, mean + 2*std, color='red', alpha=0.2, label='±2 STD')
    
    # Add title and legend
    ax.set_title(f'{species_name} Flipper Length KDE')
    ax.legend()

# Create a row of three plots for each species
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Get unique species from the dataset
species_list = penguins['species'].unique()

# Loop through species and plot each KDE
for i, species in enumerate(species_list):
    species_data = penguins[penguins['species'] == species]
    plot_kde_with_annotations(axs[i], species_data, species)

# Adjust layout
plt.tight_layout()

````

Descption of the three diagram:

Box Plot:

	•	Summarizes data with five key statistics (min, Q1, median, Q3, max).
	•	Good for comparing medians and spread.
	•	Lacks detailed information about the distribution shape and density between percentiles.

Histogram:

	•	Displays frequency distribution using bins.
	•	Shows distribution shape (skewness, modality, spread).
	•	Easy to interpret but sensitive to bin size, which can affect how data characteristics are visualized.

Kernel Density Estimator:

	•	Smoothes data into a continuous curve to estimate the probability density function.
	•	Better for showing overall distribution shape, but sensitive to bandwidth choice, which impacts detail visibility.

ChatBot explnation of the mean and median and right and left skewness

Mean
The mean is what most people call the "average." To find the mean, add up all the numbers in a set and then divide by how many numbers there are.

Median
The median is the middle number in a list of numbers arranged from smallest to largest. If there’s an odd number of values, it’s the one right in the center. If there’s an even number, it’s the average of the two middle numbers.

1. Right Skewness (Positive Skewness)
- Definition: A distribution is right-skewed if it has a long tail on the right side.
- Relationship: In a right-skewed distribution, the mean is typically greater than the median.
- Reason: The presence of high-value outliers "pulls" the mean to the right. However, the median, which is the middle value, remains relatively unaffected by these extreme values. Since the mean incorporates all values (including the high outliers), it shifts toward the right.
- Example: Incomes in a population are often right-skewed. A few people with extremely high incomes can increase the mean substantially, whereas the median remains closer to the typical (or most common) income.

2. Left Skewness (Negative Skewness)
- Definition: A distribution is left-skewed if it has a long tail on the left side.
- Relationship: In a left-skewed distribution, the mean is typically less than the median.
- Reason: Low-value outliers "pull" the mean to the left. While the mean is influenced by these extreme lower values, the median remains relatively stable because it depends only on the middle values. As a result, the mean ends up being less than the median.
- Example: Test scores in an easy exam could be left-skewed if most students score very high but a few score extremely low, dragging the mean downward.

Right skew: Mean > Median

Left skew: Mean < Median

````python
from scipy import stats
import pandas as pd
import numpy as np
import plotly.express as px

# Generating a right-skewed distribution (positive skewness)
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Right-Skewed Distribution")
fig1.show(renderer="png")

# Calculating mean and median for right-skewed distribution
mean1 = sample1.mean()
median1 = np.quantile(sample1, 0.5)

print(f"Right-Skewed Distribution: Mean = {mean1}, Median = {median1}")

# Generating a left-skewed distribution (negative skewness)
sample2 = -stats.gamma(a=2, scale=2).rvs(size=1000)
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data", title="Left-Skewed Distribution")
fig2.show(renderer="png")

# Calculating mean and median for left-skewed distribution
mean2 = sample2.mean()
median2 = np.quantile(sample2, 0.5)

print(f"Left-Skewed Distribution: Mean = {mean2}, Median = {median2}")
````

````python
import plotly.express as px
import pandas as pd

url = "https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv"
bn = pd.read_csv(url)

bn['name'] = bn['name'] + " " + bn['sex']
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)
bn = bn.sort_values(['name', 'year'])

bn['percent change'] = bn['percent'].diff()
new_name = [True] + list(bn['name'][:-1].values != bn['name'][1:].values)
bn.loc[new_name, 'percent change'] = bn.loc[new_name, 'percent']

bn = bn.sort_values('year')
bn = bn[bn['percent'] > 0.001]
fig = px.scatter(
    bn, x='percent change', y='rank',
    animation_frame='year', animation_group='name',
    size='percent', color='sex', hover_name='name',
    size_max=50, range_x=[-0.005, 0.005]
)
fig.update_yaxes(autorange='reversed')
fig.show(renderer="png")
````
